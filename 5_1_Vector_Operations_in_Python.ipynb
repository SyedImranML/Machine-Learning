5.1 Vector Operations - in Python

```python
# Importing numpy library useful in making arrays
# vertor is nothing but array in programing
# this arrays will act as vector

import numpy as np
```

```python
# we need to represent this vectors into plots
# so we need this matplotlib library
# importing matplotlib library

import matplotlib.pyplot as plt
```

```python
# importing seaborn library as it is also data visualization library

import seaborn as sns
```

```python
# getting default grids by using sns.set()
sns.set()
```

Plotting a vector

```python
plt.quiver(0,0,4,5)
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles='xy',scale=1)
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.show()
```

```python
# Changing color to blue

plt.quiver(0,0,4,5, scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Changing color to blue

plt.quiver(0,0,4,5, scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,-6, scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 + vector_2
print(sum)
```

```python
# Representation with Graph

plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,5,1, scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of two Vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

subtraction = vector_1 - vector_2
print(subtraction)
```

```python
# Representation with Graph

plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,-1,5, scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector by a scalar

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = 2*vector_1

print(vector_2)
```

```python
# Representation with Graph

plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,4,6, scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = -0.5*vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-1,-1.5, scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
