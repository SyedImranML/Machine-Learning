#### Seaborn:

* Data Visualization library

seaborn is another library for data visualization purpose so in machine learning we often deal with thousands or even lakhs of data and it is impossible to derive meaning or insights looking at the raw datasets 
but when we plot those data in a suitable dia it helps us to give some meaning that we want, this is where data visualization comes into picture.

```python
# importing the libraries

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note : seaborn has some built-in datasets 

    1. tip datasets

```python
# total bill vs. tip dataset

tips=sns.load_dataset('tips')
```

```python
tips.head(10)
```

```python
# Visualize the data

sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# setting a theme for the plots

sns.set_theme()
```

```python
# Visualize the data

sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

This is the advantage of seaborn over a matplotlib, so in matplotlib we need to mention the smoker to be different shaded color marker all you need to mention manually when you are using matplotlib.

But when you are using seaborn you no need to do that seaborn automatically find those difference and it will plotted based on that.

#### House price dataset

this dataset is not avaiable in seaborn so we need to import from sklearn.

```python
from sklearn.datasets import load_boston
```

```python
house_boston = load_boston()
```

```python
house = pd.DataFrame(house_boston.data,columns=house_boston.feature_names)
```

```python
house['Price']=house_boston.target
```

```python
house.head()
```

#### Distribution Plot

```python
sns.displot(house['Price'])
```

```python
sns.distplot(house['Price'])
```

#### Correlation:

1. Positive correlation
2. Negative correlation

#### Heat Map

```python
correlation = house.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
