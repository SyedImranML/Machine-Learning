#### Matplotlib in Python

* Matplotlib is very useful for making plots & graphs.
* Often in machine learning you will deal with lot of data and it is not possible to drive meaningful   by looking at this raw data but when we plot this data in plots & graphs its gives important   
  insights from this data this is where matplotlib library comes into picture.
  
  Useful for making plots & graphs.

```python
# importing matplotlib library

import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots

import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

It will take x as angleand it find sine of all those angles

#### plotting the data

```python
# sine wave

plt.plot(x,y)
plt.show()
```

As you can see here we have got a sine wave because x is the values between 0 & 10 and y is the sine value of all this values of x.

If we don't mention what kind of plot, it will give us a line plot so all the points will be plotted and it will be joined by a line.

```python
# cosine wave

plt.plot(x,z)
plt.show()
```

```python
# adding title , x-aixs , y-axis labels

plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine wave')
plt.title('Sine wave')
plt.show()
```

#### Multiple lines in a single plot

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

#### Bar plot

Bar plot is also very importantin data science & Machine learning.

```python
fig = plt.figure()
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Here [0,0,1,1] - 0,0 is the coordinate and 1,1 is height & width

```python
fig = plt.figure()
ax=fig.add_axes([0,0,2,2])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

In several cases in data science & machine learning we use this bar graphs to understand about the data so this is known as Data visualization.

Data visualization uses several plots & analysis purpose.

#### Pie Chart

pie chart is very useful to find the distribution of the data in the entire datasets.

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%.1f%%')
plt.show()
```

#### Scatter Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

* scatter plot is used in clustering Application.
* As you can see here in asingle graph this is scatter plot where datapoints will be scatter.
* they will be no line joining this points.

#### 3D scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z= 20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```

```python
ax = plt.figure().add_subplot(projection='3d')
# Plot a sin curve using the x and y axes.
x = np.linspace(0, 1, 100)
y = np.sin(x * 2 * np.pi) / 2 + 0.5
ax.plot(x, y, zs=0, zdir='z', label='curve in (x, y)')

# Plot scatterplot data (20 2D points per colour) on the x and z axes.
colors = ('r', 'g', 'b', 'k')

# Fixing random state for reproducibility
np.random.seed(19680801)

x = np.random.sample(20 * len(colors))
y = np.random.sample(20 * len(colors))
c_list = []
for c in colors:
    c_list.extend([c] * 20)
# By using zdir='y', the y value of these points is fixed to the zs value 0
# and the (x, y) points are plotted on the x and z axes.
ax.scatter(x, y, zs=0, zdir='y', c=c_list, label='points in (x, z)')

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Customize the view angle so it's easier to see that the scatter points lie
# on the plane y=0
ax.view_init(elev=20., azim=-35)

plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    """
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    """
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for m, zlow, zhigh in [('o', -50, -25), ('^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
